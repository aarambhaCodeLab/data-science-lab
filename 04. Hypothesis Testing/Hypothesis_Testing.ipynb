## Hypothesis Testing

**Hypothesis testing** is a statistical method that helps us make decisions about a population using sample data.

**Null Hypothesis:** (H₀) The "default" assumption — no effect, no difference.
Alternative Hypothesis: What we are trying to prove — there IS an effect or difference.

### Significance Level (α)

The **significance level (α)** is the threshold we set for how much risk we're willing to take of making a wrong decision.

## Z-Test Example — E-Commerce Website Visitors

### Problem Statement
An e-commerce platform claims it receives an average of **100 visitors per day** (this is the known population mean).
We collected data for **38 days** and want to check: *Is the actual average significantly different from 100?*
**significance level α = 0.05**.


## Decision Rule

### Reject the Null Hypothesis (H₀)

- If **Z₍calculated₎** lies **outside** the range  
  \[
  [-Z_{critical},\; Z_{critical}]
  \]
- The sample mean is **significantly different** from the population mean.
- This means there is **sufficient evidence** to conclude that the observed difference  
  is **unlikely due to random sampling alone**.

---

### Fail to Reject the Null Hypothesis (H₀)

- If **Z₍calculated₎** lies **within** the range  
  \[
  [-Z_{critical},\; Z_{critical}]
  \]
- The sample mean is **not significantly different** from the population mean.
- This means the observed difference **could reasonably be due to random sampling**.


## When to Use Z-test and T-test

### Use a Z-test when all of the following conditions are satisfied:
- Population standard deviation (σ) is **known**
- Sample size is **large** (usually \( n \geq 30 \))
- Data is **normally distributed**, or the sample size is large enough for the  
  **Central Limit Theorem (CLT)** to apply

---

### Use a T-test when any of the following conditions are true:
- Population standard deviation (σ) is **unknown**
- Sample size is **small** (usually \( n < 30 \))
- Population is **approximately normal**


```python
# Known population parameters
pop_mean = 100   # Population mean
pop_sd   = 15    # Population standard deviation (known)
alpha    = 0.05  # Significance level

# Sample data: daily visitors over 38 days
visitors = [
    117, 119, 132, 106, 133, 121, 102, 118, 103, 103,
    113, 81,  84,  101, 94,  114, 96,  88,  131, 106,
    111, 88,  101, 111, 92,  115, 100, 105, 100, 137,
    109, 94,  122, 91,  113, 80,  90,  112
]

# Compute sample statistics
sample_mean = np.mean(visitors)
sample_size = len(visitors)

print(f"Sample Size     : {sample_size}")
print(f"Sample Mean     : {sample_mean:.4f}")
print(f"Population Mean : {pop_mean}")
print(f"Population SD   : {pop_sd}")
```

```python
# Calculate the Z-score
z_score_calculated = (sample_mean - pop_mean) / (pop_sd / np.sqrt(sample_size))

# Critical Z value for two-tailed test at alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)

print(f"Calculated Z-Score  : {z_score_calculated:.4f}")
print(f"Critical Z-Value    : ±{z_critical:.4f}")
print(f"Rejection Region    : Z < -{z_critical:.4f}  or  Z > {z_critical:.4f}")
```

```python
p_value_z = 2 * (1 - stats.norm.cdf(abs(z_score_calculated)))

print("         Z-TEST RESULTS SUMMARY")

print(f"  Calculated Z-Score : {z_score_calculated:.4f}")
print(f"  Critical Z-Value   : ±{z_critical:.4f}")
print(f"  P-Value            : {p_value_z:.4f}")
print(f"  Significance Level : {alpha}")
print(" ==============================================")

if abs(z_score_calculated) > z_critical:
    print("  DECISION: REJECT H₀")
    print("  The sample mean IS significantly different")
    print("  from the population mean.")
    
else:
    print("  DECISION: FAIL TO REJECT H₀")
    print("  The sample mean is NOT significantly different")
    print("  from the population mean.")
```

**T-test**

**To test weather the average daily study time of students is significantly different from the expected standard of 4 hrs per day**

```python
import numpy as np

study_hours = [3.5, 4.2, 3.8, 4.6, 5.0, 3.9, 4.1, 4.8, 3.7, 4.3]

expected_mean = 4 # Expected study time (hours)

sample_mean = np.mean(study_hours)
sample_std = np.std(study_hours, ddof=1)
n = len(study_hours)

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Sample Size:", n)
```

```python
# Performing one sample t test
t_stat, p_value= stats.ttest_1samp(study_hours, expected_mean)
print("T-statistics:", t_stat)
print("P-value:", p_value)
alpha=0.05

if p_value < alpha:
    print("Reject Ho : Study time is significantly different from 4 hours")
else:
    print("Fail to reject Ho : Study time is not significantly different from 4 hours")
```
