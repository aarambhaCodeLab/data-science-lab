## Problem Statement

A retailer supermarket wants to evaluate its **monthly sales performance** to understand **customer behavior**, **seasonal trends**, and the **influence of discounts on revenue**.

As a **data analyst**, you are required to **visualize the dataset** and **extract meaningful insights** that can assist management in planning **promotions**, **inventory management**, and **sales strategies**.


```python
# Setting white grids for clear appearence
sns.set_theme(style="whitegrid")

data = {
    "Month" : ["Jan" , "Feb" , "Mar" , "Apr" , "May" , "Jun" , "Jul" , "Aug" , "Sep" , "Oct" , "Nov" , "Dec"],
    "Sales" : [120000 , 135000 , 150000 , 145000 , 160000 , 170000 , 180000 , 175000 , 165000 , 190000 , 210000 , 250000],
    "Customers" : [420 , 450 , 470 , 460 , 480 , 500 , 520 , 510 , 490 , 530 , 560 , 600],
    "Discount" : [5 , 5 , 10 , 8 , 10 , 12 , 15 , 12 , 10 , 8 , 5 , 5]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df["Month"], df["Sales"], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales (NPR)")
plt.grid(True)
plt.show()
```

```python
#BAR CHART 
plt.figure(figsize=(8,5))
plt.bar(df["Month"],df["Sales"], color="blue", alpha=0.6)
plt.title("Monthly sales trend")
plt.xlabel("Month")
plt.ylabel("Sales(NPR)")
plt.grid(False)
plt.show()
```

# Heat Map

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[["Sales", "Customers","Discount"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heat Map")
plt.show()
```

# Box Plot (Sales Distribution)

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df["Sales"], color="red")
plt.title("Sales Distribution and OUtlier Detection")
plt.show()
```

# Bubble Chart 

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="Sales",
    y="Customers",
    size="Discount",
    hue="Discount",
    sizes=(50, 800),
    alpha=0.7
)

plt.title("Bubble Chart: Sales vs Customers (Discount Impact)")
plt.xlabel("Sales (NPR)")
plt.ylabel("Number of Customers")
plt.grid(True)
plt.show()
```

# Pair Plot 

```python
sns.pairplot(df[["Sales","Customers","Discount"]])
plt.show()
```

```python
sns.countplot(x="Discount", data=df)
plt.title("Discount Frequency")
plt.show()
```

```python
# Season mapping
season_map = {
    "Winter": ["Dec","Jan","Feb"],
    "Spring": ["Mar","Apr","May"],
    "Summer": ["Jun","Jul","Aug"],
    "Autumn": ["Sep","Oct","Nov"]
}

# Group sales by season
season_sales = {}
for season, months in season_map.items():
    season_sales[season] = df[df["Month"].isin(months)]["Sales"].sum()

season_df = pd.Series(season_sales)

# Pie chart
plt.figure()
plt.pie(season_df, labels=season_df.index, autopct='%1.1f%%')
plt.title("Sales by Season")
plt.show()
```
