```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Define the population (example: normally distributed population with mean 50, std 10)
population_size = 10000
population_mean = 50
population_std = 10
population = np.random.normal(population_mean, population_std, population_size)

# Define sample sizes and number of samples
sample_sizes = [10, 30, 50, 100, 200, 500]
num_samples = 1000

# Create the plot
plt.figure(figsize=(14, 10))

for i, size in enumerate(sample_sizes, start=1):
    sample_means = [
        np.mean(np.random.choice(population, size=size, replace=False))
        for _ in range(num_samples)
    ]

    plt.subplot(3, 2, i)
    sns.histplot(sample_means, bins=30, kde=True, color="red", alpha=0.7)

    plt.title(f"Sample Mean Distribution (n = {size})", fontsize=14)
    plt.xlabel("Sample Mean", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)

plt.suptitle("Demonstration of Central Limit Theorem", fontsize=18, y=1.02)
plt.tight_layout()
plt.show()
```


### 1. The Shape Becomes a Symmetrical Bell Curve
At a small sample size ($n = 10$), the histogram looks a bit rough and slightly uneven. But as the sample size increases to $30$, $50$, and beyond, the red line smooths out completely.

No matter what the original population looks like (even if it is highly skewed), the distribution of the sample means will always turn into a symmetrical **Normal Distribution** (bell curve) as the sample size grows. This transformation reliably locks in once the sample size reaches $n \ge 30$.

### 2. The Center Always Stays the Same
 Look at the peak of the bell curve on the bottom axis for every single plot. Whether the sample size is small ($n = 10$) or large ($n = 500$), the center of the curve stays exactly at **50.0**.

The average of all your sample means will always target the true center of the actual population. Taking larger samples doesn't move the center; it just makes our estimate more accurate.
